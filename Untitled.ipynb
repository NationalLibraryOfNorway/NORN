```python
import dhlab as dh
import pandas as pd
import re
```

```python
imag = pd.read_excel('ImagiNation_Clean_3.xlsx')
```

```python
konsensus = pd.read_excel('konsensuskorpus.xlsx')
```

```python
def text2urn(t):
    res = re.findall("URN:NBN:no-nb[^ ]+", t)
    if res == []:
        res = ""
    else:
        res = res[0]
    return res
```

```python
konsensus['urn'] = konsensus['oaiid/URN'].apply(text2urn)
```

```python
kons = dh.Corpus()
```

```python
kons.extend_from_identifiers(konsensus.urn)
```

```python
kons.frame["dhlabid title authors city year publisher langs subjects ddc ocr_creator urn".split()].to_csv('corpus_konsensus.csv')
```
