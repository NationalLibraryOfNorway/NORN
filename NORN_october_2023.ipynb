```python
tot = combo[["1800", "1840", "NatRom"]].sum(axis = 1)
```

```python
def make_coll(corpus, words):
    coll = dh.Collocations(corpus,words, before = 5, after = 5).frame
    coll.sort_values(by = 'counts', ascending=False)
    coll['diff'] = (coll['counts']/coll['counts'].sum())/(tot/tot.sum())
    return coll
```
