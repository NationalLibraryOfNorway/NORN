```python
import pandas as pd
from bs4 import BeautifulSoup
import os
```

```python

def get_xml_data(folder):
    _,_,fs = next(os.walk(folder))
    res = []
    for f in [f for f in fs if f.endswith('xml')]:
        try:
            with open(os.path.join(folder, f)) as fp:
                soup = BeautifulSoup(fp, 'lxml')
                res.append(get_table_data(soup))
        except:
            print("problemer med", f)
    return res

def get_table_data(soup):
    """Parse old xml file"""
    
    # pick out the top level fields with metadata
    
    top_fields = "work author genre year publication publisher".split()
    res = {tf:soup.find(tf).text for tf in top_fields}

    # each row has a set of subfields
    
    fields = "author year pages period urn notes".split()
    n =  soup.find('citation_sources')
    res['table'] = []
    for z in n.find_all('source'):
        row = {field:z.find(field).text for field in fields}
        res['table'].append(row)
        
    return res
```

```python
norn = get_xml_data("Nasjonalromantikk/")
```
